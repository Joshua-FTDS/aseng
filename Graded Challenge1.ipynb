```python
class cartitem: 
    def __init__ (self, nama, harga):
        self.nama = nama
        self.harga = harga

class shoopingcart:
    def __init__ (self):
        self.container = []

    def tambah_barang(self, benda):
        self.container.append(benda)
        print(f"Barang {benda.nama} berhasil dimasukkan ke keranjang.")

    def hapus_barang(self, barang_yang_dicari):
        if len(self.container) == 0:
            print("keranjang masih kosong")
        else :
            for benda in self.container:
                if barang_yang_dicari == benda.nama:
                    self.container.remove(benda)
                    print(f"Barang {benda.nama} berhasil dihapus dari keranjang.")
                    break
                else:
                    print("Barang tidak ada")

    def tampilkan_barang(self):
        print("Barang di keranjang:")
        if len(self.container) == 0:
            print("keranjang masih kosong")
        i = 1
        for benda in self.container:
            print(f'{i}. {benda.nama} | {benda.harga}')
            i = i + 1

    def total_cart(self):
        total = 0
        for x in self.container:
            total += x
            return (total)

belanjaan = shoopingcart()
run = True
while run:
    print('''\nSelamat Datang di Keranjang Belanja Toko Makmur!

Menu:
1. Menambah Barang
2. Hapus Barang
3. Tampilkan Barang di Keranjang
4. Lihat Total Belanja
5. Exit''')
    
    menu = input("pilih menu")

    if menu == "1":
        nama = input("Masukkan nama barang")
        harga = input("Masukkan harga")

        benda = cartitem(nama, harga)
        belanjaan.tambah_barang(benda)

    elif menu == "2":
        nama_benda = input("Masukkan nama barang")
        belanjaan.hapus_barang(nama_benda)

    elif menu == "3":
        belanjaan.tampilkan_barang()

    elif menu == "4":
        #belanjaan.total_cart()
        print(f"Total belanja anda: {belanjaan.total_cart}")

    elif menu == "5":
        print("Sampai Jumpa! Terima kasih sudah belanja di Toko Makmur.")
        run = False
```
